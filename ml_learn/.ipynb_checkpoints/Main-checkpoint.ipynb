```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load dataset
# -----------------------------
data = pd.read_csv("/home/pythonic/Desktop/FYP/project/learn_bpf/extract_sysCalls_to_map/dataset/parsing_data/finalTest.csv")
X = data.drop("LABEL", axis=1)
y = data["LABEL"]

# -----------------------------
# 2. Add small noise to numeric features
# -----------------------------
noise_level = 0.02  # 2% noise
numeric_cols = X.select_dtypes(include=np.number).columns
X_noisy = X.copy()
for col in numeric_cols:
    X_noisy[col] += np.random.normal(0, noise_level * X_noisy[col].std(), size=X_noisy.shape[0])

# -----------------------------
# 3. Slightly overlap classes by flipping a small % of labels (optional)
# -----------------------------
y_mod = y.copy()
n_flip = int(0.02 * len(y_mod))  # flip 2% of labels
flip_indices = np.random.choice(y_mod.index, n_flip, replace=False)
y_mod.iloc[flip_indices] = 1 - y_mod.iloc[flip_indices]

# -----------------------------
# 4. Train-test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_noisy, y_mod, test_size=0.2, random_state=42, stratify=y_mod
)

# -----------------------------
# 5. Scale features
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# 6. SMOTE oversampling
# -----------------------------
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

# -----------------------------
# 7. RandomForest training
# -----------------------------
rf = RandomForestClassifier(random_state=42, max_depth=7, min_samples_leaf=3)
rf.fit(X_train_res, y_train_res)

# -----------------------------
# 8. Predictions
# -----------------------------
y_pred = rf.predict(X_test_scaled)
y_proba = rf.predict_proba(X_test_scaled)[:, 1] if hasattr(rf, "predict_proba") else None

# -----------------------------
# 9. Evaluation metrics
# -----------------------------
print("===== Classification Metrics =====")
print(f"Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred, zero_division=0):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred, zero_division=0):.4f}")
print(f"F1-score : {f1_score(y_test, y_pred, zero_division=0):.4f}")
if y_proba is not None:
    print(f"ROC-AUC  : {roc_auc_score(y_test, y_proba):.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Classification report
print("\n===== Classification Report =====")
print(classification_report(y_test, y_pred, zero_division=0))

# -----------------------------
# 10. Feature Importances
# -----------------------------
fi_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\n===== Feature Importances =====")
print(fi_df)

plt.figure(figsize=(10,6))
sns.barplot(x="Importance", y="Feature", data=fi_df, color="mediumslateblue")
plt.title("Feature Importances")
plt.tight_layout()
plt.show()

# -----------------------------
# 11. Cross-validation
# -----------------------------
cv_scores = cross_val_score(rf, X_noisy, y_mod, cv=5)
print("\n5-fold cross-validation scores:", cv_scores)
print("Mean CV score:", cv_scores.mean())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType
from skl2onnx.common.data_types import DoubleTensorType

# -----------------------------
# 1. Load dataset
# -----------------------------
data = pd.read_csv("/home/pythonic/Desktop/FYP/project/learn_bpf/extract_sysCalls_to_map/dataset/parsing_data/output.csv")
X = data.drop("LABEL", axis=1)
y = data["LABEL"]

# -----------------------------
# 2. Add small noise to numeric features
# -----------------------------
noise_level = 0.02  # 2% noise
numeric_cols = X.select_dtypes(include=np.number).columns
X_noisy = X.copy()
for col in numeric_cols:
    X_noisy[col] += np.random.normal(0, noise_level * X_noisy[col].std(), size=X_noisy.shape[0])

# -----------------------------
# 3. Slightly overlap classes by flipping a small % of labels (optional)
# -----------------------------
y_mod = y.copy()
n_flip = int(0.02 * len(y_mod))  # flip 2% of labels
flip_indices = np.random.choice(y_mod.index, n_flip, replace=False)
y_mod.iloc[flip_indices] = 1 - y_mod.iloc[flip_indices]

# -----------------------------
# 4. Train-test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_noisy, y_mod, test_size=0.2, random_state=42, stratify=y_mod
)

# -----------------------------
# 5. Scale features
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# 6. SMOTE oversampling
# -----------------------------
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

# -----------------------------
# 7. RandomForest training
# -----------------------------
rf = RandomForestClassifier(random_state=42, max_depth=7, min_samples_leaf=3)
rf.fit(X_train_res, y_train_res)

# -----------------------------
# 8. Predictions
# -----------------------------
y_pred = rf.predict(X_test_scaled)
y_proba = rf.predict_proba(X_test_scaled)[:, 1] if hasattr(rf, "predict_proba") else None

# -----------------------------
# 9. Evaluation metrics
# -----------------------------
print("===== Classification Metrics =====")
print(f"Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred, zero_division=0):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred, zero_division=0):.4f}")
print(f"F1-score : {f1_score(y_test, y_pred, zero_division=0):.4f}")
if y_proba is not None:
    print(f"ROC-AUC  : {roc_auc_score(y_test, y_proba):.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Classification report
print("\n===== Classification Report =====")
print(classification_report(y_test, y_pred, zero_division=0))

# -----------------------------
# 10. Feature Importances
# -----------------------------
fi_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf.feature_importances_
}).sort_values(by="Importance", ascending=False)
print("\n===== Feature Importances =====")
print(fi_df)
plt.figure(figsize=(10,6))
sns.barplot(x="Importance", y="Feature", data=fi_df, color="mediumslateblue")
plt.title("Feature Importances")
plt.tight_layout()
plt.show()

# -----------------------------
# 11. Cross-validation
# -----------------------------
cv_scores = cross_val_score(rf, X_noisy, y_mod, cv=5)
print("\n5-fold cross-validation scores:", cv_scores)
print("Mean CV score:", cv_scores.mean())

# =============================
# 12. EXPORT TO ONNX
# =============================
print("\n===== Exporting Model to ONNX =====")

# Create a pipeline with the trained scaler and model
# Note: We use the scaler that was fit on X_train and the RF trained on SMOTE data
pipeline = Pipeline([
    ('scaler', scaler),
    ('classifier', rf)
])

# Define the input type (31 features, float32)
initial_type = [('float_input', DoubleTensorType([None, 31]))]

# Convert to ONNX with opset version 12 (widely supported)
onnx_model = convert_sklearn(
    pipeline,
    initial_types=initial_type,
    target_opset=12,
    options={id(rf): {'zipmap': False}}  # Remove ZipMap for cleaner output
)

# Save the model
with open("rf_pipeline.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())

print("Model exported successfully to rf_pipeline.onnx!")

# Verify the export worked
import onnxruntime as ort
sess = ort.InferenceSession("rf_pipeline.onnx")
print("\nONNX Model Info:")
print("Input name:", sess.get_inputs()[0].name)
print("Input shape:", sess.get_inputs()[0].shape)
print("Output names:", [output.name for output in sess.get_outputs()])

# Test the ONNX model with the same test data
print("\n===== Testing ONNX Model =====")
onnx_pred = sess.run(None, {sess.get_inputs()[0].name: X_test_scaled.astype(np.float64)})
onnx_labels = onnx_pred[0]
onnx_probas = onnx_pred[1]

print(f"ONNX Accuracy: {accuracy_score(y_test, onnx_labels):.4f}")
print("ONNX and scikit-learn predictions match:", np.array_equal(y_pred, onnx_labels))
```
